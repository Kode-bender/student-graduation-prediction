***assign x to independent variable so we exclude "Target" column because its a dependent variable.***

```python
x = stud_dataset.drop('Target', axis = 1)
```

***assign the dependent variable***

```python
y = stud_dataset['Target']
```

***import train_test_split from sklearn library to split training and testing dataset(80% : 20%)***

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state= 40)
```

***importing StandardScaler library from sklearn to help in feature scaling of the dataset, which will categories same features in the same scale***

```python

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test= sc.transform(x_test)

x_test
```

```python
import numpy as np

value_counts = y.value_counts()
value_counts
```

***convert to array***

```python
y_train=np.array(y_train)
y_test=np.array(y_test)
```

Importing different classification models to choose the best fit

```python

from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

***Training LogisticRegression model and evaluating it's performance on this dataset***

```python
logR = LogisticRegression()
logR.fit(x_train,y_train)
y_prediction1= logR.predict(x_test)
print(accuracy_score(y_test, y_prediction1))
```

*Training KNeigborsClassifier model and evaluating it's performance on this dataset*

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_prediction3 = knn.predict(x_test)
print(accuracy_score(y_test, y_prediction3))
```

***Creating pandas DataFrame, to visualize score***

```python

Models_Percentage = pd.DataFrame({'Models':['LogisticRegression',
                                            'KNeighborsClassifier'], 'AccuracyScore':[accuracy_score(y_test, y_prediction1),
                                                                                            accuracy_score(y_test, y_prediction3),
                                                                                            ]})
```

```python
Models_Percentage
```

***Visualizing accuracy score dataframe using searborn library. According to the accuravy score LogisticRegression is the best fit***

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x='AccuracyScore', y='Models', data=Models_Percentage)
```
